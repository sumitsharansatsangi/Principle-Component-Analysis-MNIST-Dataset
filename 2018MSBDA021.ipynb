```python
import pandas as pd
import numpy as np
import os
```

```python
df=pd.read_csv("/home/sumit/Downloads/digit-recognizer/train.csv")
```

```python
X=df.drop(["label"],axis=1)
label=df["label"]
```

```python
def pca(data):
    cov=np.cov(X.T)
    eig_vals,eig_vecs=np.linalg.eig(cov)
    idx=np.argsort(eig_vals)[::-1]
    sorted_eig_vals=sorted(eig_vals,reverse=True)
    sorted_eig_vecs=eig_vecs[idx]
    tot = sum(eig_vals)
    var_exp = [(i / tot)*100 for i in sorted_eig_vals]
    cum_sum= np.cumsum(var_exp)
    i=0
    while i >= 0:
        #print("Cumulative variance captured as we travel ",i,"component is :",a[i])
        if(cum_sum[i]>99):
            break
        i=i+1
    pca_vec=np.array(pd.DataFrame(eig_vecs).iloc[:,0:i])
    X_pca = X.dot(pca_vec)
    X_pca_real=pd.DataFrame(np.real(X_pca))
    return X_pca_real
```
